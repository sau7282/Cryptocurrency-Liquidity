```python
# importing libararies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
```

```python
df1 = pd.read_csv('../data/coin_gecko_2022-03-16.csv')  # First file
df2 = pd.read_csv('../data/coin_gecko_2022-03-17.csv')  # Second file

print(df1.shape)
print(df2.shape)
# df1.head()
# df2.head()
```

```python
# Concatenate the dataframes vertically (add rows)
df = pd.concat([df1, df2], axis=0, ignore_index=True)

# Check if the concatenation was successful
print(df.shape)
df.head()
```

```python
df.info()
```

```python
'''
The columns are as follows:

coin: Name of the cryptocurrency.

symbol: Symbol of the cryptocurrency (e.g., BTC, ETH).

price: Price of the cryptocurrency.

1h, 24h, 7d: Percentage change in price over the past 1 hour, 24 hours, and 7 days.

24h_volume: 24-hour trading volume.

mkt_cap: Market capitalization.

date: Date of the data entry (currently an object type, which likely needs to be converted to datetime).
'''
```

```python
df.describe()
```

```python
df.head()
```

```python
df['24h_volume'] = df['24h_volume'].apply(lambda x: '{:.2f}'.format(x))
df['mkt_cap'] = df['mkt_cap'].apply(lambda x: '{:.2f}'.format(x))
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df.columns
```

```python
# unique value for date columns
df['date'].unique()
```

```python
# change data type for date column
df['date'] = pd.to_datetime(df['date'])
```

```python
# Check and handle missing values
df.isnull().sum()
```

```python
# since there are very less number of null records, we are good to drop it.
df.dropna(inplace=True)
```

```python
# checking after dropping null values
df.isnull().sum()
```

```python
# Univariate Analysis (Distribution of Features)
```

```python
# Price distribution
sns.histplot(df['price'], bins=50, kde=True)
plt.title("Distribution of Cryptocurrency Prices")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()
```

```python
# Market Cap vs Price
sns.scatterplot(x='mkt_cap', y='price', data=df)
plt.title("Market Cap vs Price")
plt.xlabel("Market Cap")
plt.ylabel("Price")
plt.show()
```

```python
#  Correlation Heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df[['price', '1h', '24h', '7d', '24h_volume', 'mkt_cap']].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
'''
Explanation:

Histplot gives an idea of how prices are distributed (log-normal, skewed, etc.).

Scatterplot shows how price and market cap are related — useful for liquidity insights.

Heatmap highlights linear relationships that might inform feature selection.

'''
```

```python

df['mkt_cap'] = pd.to_numeric(df['mkt_cap'], errors='coerce')

# Top 10 coins by market cap (average)
top_coins = df.groupby('coin')['mkt_cap'].mean().sort_values(ascending=False).head(10)

# Plot the bar chart
top_coins.plot(kind='bar', title='Top 10 Coins by Avg Market Cap')
plt.ylabel("Average Market Cap")
plt.show()
```

```python
# Convert '24h_volume' to numeric
df['24h_volume'] = pd.to_numeric(df['24h_volume'], errors='coerce')

# Top 10 coins by 24h trading volume (average)
top_volume = df.groupby('coin')['24h_volume'].mean().sort_values(ascending=False).head(10)

# Plot the bar chart
top_volume.plot(kind='bar', color='orange', title='Top 10 Coins by Avg 24h Volume')
plt.ylabel("Average 24h Volume")
plt.show()
```

```python
# Explanation: These rankings help identify high-liquidity coins, useful for modeling and insights.
```

```python
# Ensure '24h_volume' and 'mkt_cap' are numeric
df['24h_volume'] = pd.to_numeric(df['24h_volume'], errors='coerce')
df['mkt_cap'] = pd.to_numeric(df['mkt_cap'], errors='coerce')

# Calculate liquidity ratio
df['liquidity_ratio'] = df['24h_volume'] / df['mkt_cap']

# Plot liquidity ratio distribution
sns.histplot(df['liquidity_ratio'], kde=True)
plt.title("Liquidity Ratio Distribution")
plt.xlabel("Liquidity Ratio")
plt.ylabel("Frequency")
plt.show()
```

# Explanation: liquidity_ratio (volume-to-market-cap) is a useful indicator of how liquid a coin is. This could be a target or feature in your model.
